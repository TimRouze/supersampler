```python
import pandas as pd
import numpy as np
import math
```

```python
def basic_metrics(data):
    max = np.max(data)
    min = np.min(data)
    mean = np.mean(data)
    std = np.std(data)
    median = np.median(data)
    print(f"Mean = {mean}\nStd = {std}\nMedian = {median}\nMax = {max}\nMin = {min}")
```

```python
def sum_harmo(exp, diff):
    res = 0
    for dif in diff:
        res += 1/(dif**exp)
    return res
```

```python
def moy_harmo(exp, diff, size):
    sum_h = sum_harmo(exp, diff)
    res = (size/abs(sum_h))**1/exp
    return res
```

```python
simka = pd.read_csv("expe100salmo/results_jaccard1.csv", sep = ",", header = 0)
simka
```

```python
simka = simka.to_numpy()
ltri_simka = simka[np.tril_indices_from(simka, k = -1)]
basic_metrics(ltri_simka)
```

```python
sourmash = pd.read_csv("expe100salmo/sms1024.csv", sep = ",", header = 0)
sourmash
```

```python
sourmash = sourmash.to_numpy()
ltri_sm = sourmash[np.tril_indices_from(sourmash, k = -1)]
basic_metrics(ltri_sm)
```

```python
spsp = pd.read_csv("expe100salmo/results_jaccard1024.csv", sep = ",", header = 0)
spsp
```

```python
spsp = spsp.to_numpy()

ltri_spsp = spsp[np.tril_indices_from(spsp, k = -1)]

basic_metrics(ltri_spsp)
```

```python
diff_simka_sm = abs(ltri_simka - ltri_sm)
diff_simka_spsp = abs(ltri_simka - ltri_spsp)
diff_sm_spsp = abs(ltri_sm- ltri_spsp)
norm_sm_spsp = np.linalg.norm(diff_sm_spsp,1)

norm_simka_sm = np.linalg.norm(diff_simka_sm,1)
norm_simka_spsp = np.linalg.norm(diff_simka_spsp,1)
norm_simka_sm2 = np.linalg.norm(diff_simka_sm,2)
norm_simka_spsp2 = np.linalg.norm(diff_simka_spsp,2)
print(f'Norm1 of differences between supersampler and Sourmash: {norm_sm_spsp}\n')
print(f'Norm1 of differences between Simka and Sourmash: {norm_simka_sm}\nNorm1 of differences between Simka and Supersampler: {norm_simka_spsp}')
print(f'Norm2 of differences between Simka and Sourmash: {norm_simka_sm2}\nNorm2 of differences between Simka and Supersampler: {norm_simka_spsp2}')
```
